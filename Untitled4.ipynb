```python

df = pd.read_pickle("history/tables/monthly_report.pkl")

#print(df.head())
for filename in files:

    print(filename)
    #filename='8077 洛碁_每月營收.xlsx'
    
    stock=filename[:filename.find('_')]
    #print(stock)

    #endDate = df[df.index.get_level_values('stock_id') == stock].sort_index(level='date', ascending=True).head(1).index.get_level_values('date')
    
    #print(endDate)
    

    df_reader = pd.read_excel('D:/stock_grabber/Artie_programs/每月營收/'+filename,sheet_name='每月營收')
   
    if df_reader.empty:
        continue
    
    #print(df_reader.iloc[0,0])
    #break
    endDate=df[df.index.get_level_values('stock_id') == df_reader.iloc[0,0]].head(1).index.get_level_values('date')
   
    print(endDate[0])
    
    df_reader['stock_id']=df_reader['stock_id'].fillna(df_reader.iloc[0,0])
    
    df_reader.reset_index()
    df_reader=df_reader.set_index(['stock_id','date'])
    
    df_reader.columns=['當月營收','上月比較增減(%)','去年同月增減(%)','當月累計營收','前期比較增減(%)']
    
    #print(df_reader)

    df_reader=df_reader[df_reader.index.get_level_values('date') < endDate[0]]
    #print(df_reader)
    
    if df_reader.empty:
        continue
    
    #print(df_reader.shape)
    #print(df.shape)
    #print(df[df.index.get_level_values('stock_id') == stock].shape)
    #print(df_reader.index.get_level_values('stock_id'))
    df=df.append(df_reader,sort=False).sort_index(level='date', ascending=True)
    #print(df.shape)
    #print(df[df.index.get_level_values('stock_id') == stock])
    #print(df)
    
```

```python
df=df.sort_index(level='stock_id', ascending=True)
#print(df)
to_pickle(df, 'monthly_report_new')
```
